Import MLPRegressor for model, train_test_split to split data, StandardScaler to scale input variables, r2_score and mean_absolute_error for model metrics

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

Create input and output sets and split them into test and train groups. Scale input set.

```python
x=df.iloc[:,:-1]
x = x.iloc[: , 1:]
y=df.iloc[:,-1:]
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=1, test_size=0.2)
sc_x = StandardScaler()
x_trainscaled=sc_x.fit_transform(x_train)
x_testscaled=sc_x.transform(x_test)
```

Create model using training data sets

```python
reg = MLPRegressor(hidden_layer_sizes=(64,64,64),verbose='true',activation="relu" ,random_state=1, max_iter=2000).fit(x_trainscaled, y_train.values.ravel())
```

Evaluate model using testing data set

```python
y_pred=reg.predict(x_testscaled)
print("R2 Score of ", (r2_score(y_pred, y_test)))
print("MAE Score of ", (mean_absolute_error(y_pred, y_test)))
```
